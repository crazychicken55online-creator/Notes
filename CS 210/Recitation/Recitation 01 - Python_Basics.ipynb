
## 3) Collections: Lists, Tuples, Sets, Dicts


```python
# Examples â€” run and observe
nums = [3, 1, 4, 1, 5]
nums.append(9)
nums.sort()
print("list:", nums)

# Tuples (ordered, immutable)
pt = (10, 20)
print("tuple:", pt)

# Sets (unordered, unique elements)
colors = {"red", "blue", "red", "green"}
colors.add("yellow")
print("set:", colors)

# Dicts (key -> value)
student = {"name": "Ada", "id": 101, "skills": ["python", "pandas"]}
student["grad_year"] = 2026
print("dict:", student)
```


**Exercise 3:**  
1. Start with `scores = [88, 92, 75, 92, 100]`.  
   - Add a new score `85`, then sort ascending.  
   - Remove duplicates while preserving order (hint: use a loop or dict.fromkeys).  
2. Create a dict `person` with keys: `name`, `age`, `hobbies` (list of 2+ strings).  
   - Append a new hobby, then print a sentence using fâ€‘string with name and hobby count.


```python
# Your work for Exercise 3
scores = [88, 92, 75, 92, 100]
scores.append(85)
scores.sort()
res = []
for i in range(len(scores)-1):
    if scores[i] != scores[i+1]:
        res.append(scores[i])
res.append(scores[-1])
print(res)
```


## 4) Control Flow: if/elif/else, loops, comprehensions


```python
# Examples â€” run and observe
n = 17
if n % 2 == 0:
    kind = "even"
elif n % 3 == 0:
    kind = "divisible by 3"
else:
    kind = "odd and not divisible by 3"
print(kind)

# loops
items = ["book", "laptop", "pen"]
for i, item in enumerate(items, start=1):
    print(i, item)

# list comprehension
squares = [k*k for k in range(1, 8)]
print("squares:", squares)
```


**Exercise 4:**  
Write a loop that prints numbers **1â€“20**, but:
- Print `"Fizz"` for multiples of 3,
- `"Buzz"` for multiples of 5,
- `"FizzBuzz"` for multiples of 15,
- Otherwise print the number.
(Yes, the classic ðŸ’¥)


```python
# Your work for Exercise 4
for i in range(1, 21):
    if i % 15 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)

```
